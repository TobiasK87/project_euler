```python
def primes(n):
    r = [True] * n
    r[0] = r[1] = False 
    r[4::2] = [False] * len(r[4::2])
    for i in range(3,int(1 + n**0.5),2):
        if r[i]:
            r[i*i::2*i] = [False] * len(r[i*i::2*i])
    return r

prime_list = primes(10**7)
prime_list = [i for i,x in enumerate(prime_list) if x]
prime_list[:5]
```

```python
def is_prime(n):
    """
    Assumes that n is a positive natural number
    """
    if n == 1:
        return False
    for i in prime_list:
        if i**2>n:
            break
        if n % i == 0:
            return False
    return True
```

```python
def isRightTrunc(n):
    str_n = str(n)
    for i in range(len(str_n)-1):
        substr = n
        if i>=1:
            substr = int(str_n[:-1*i])
        # print(substr, substr%(sum([int(a) for a in str(substr)])))
        if substr%(sum([int(a) for a in str(substr)]))!=0:
            return False
    return True        
```

```python
def isStrongRightTrunc(n):
    if isRightTrunc(n):
        k = int(n/(sum([int(a) for a in str(n)])))
        print(k)
        return is_prime(k)
    return False
```
