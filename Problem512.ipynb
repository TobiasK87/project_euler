```python
# Vorüberlegungen:
# Es ist recht einfach zu beweisen, dass die Summe von i=1 bis n über phi(n^i) (mod n+1) = 0 für n gerade und = phi(n) für n ungerade ist.
# Es bleibt also die Summe über alle ungeraden n bis 5*10**8 über phi(n) zu berechnen.
# Es gibt bestimmt elegantere Wege, aber der unten implementierte Weg berechnet mit einem Einschluss-Ausschluss-Verfahren
# phi(n) ist ja gerade die Anzahl aller i<=n, die teilerfremd zu n sind
# Das wären somit maximal 1+3+5+7+9+...+(5*10**8-1)
# Aber natürlich müssen alle Vielfachen jeder Primzahl wieder abgezogen werden (Auschluss)
# Dann schließt man aber teilweise zu viel aus (beispielsweise Vielfache von 3 UND von 5 (also von 15) gleichzeitig), daher muss man die wieder addieren (Einschluss)
# usw.
# Das muss nur über paarweise verschiedene Primzahlen gemacht werden. Da 3*5*7*11*13*17*19*23 noch kleiner als 5*10**8 ist, aber 3*5*7*11*13*17*19*23*29 größer, müssen bis zu 8-fache Primzahlkombinationen beim Einschluss-Ausschluss-Verfahren berücksichtigt werden
```

```python
def primes(n):
    r = [True] * n
    r[0] = r[1] = False 
    r[4::2] = [False] * len(r[4::2])
    for i in range(3,int(1 + n**0.5),2):
        if r[i]:
            r[i*i::2*i] = [False] * len(r[i*i::2*i])
    return r

def numberToReduce(n,k):
    t = n//k
    t = t - ((t+1)%2)
    return int(((t+1)/2)**2)
def numberToIncrease(n,k):
    a = (n-1)//k
    return int(a*(a+1)/2)

prime_list = primes(5*10**8)
prime_list = [i for i,x in enumerate(prime_list) if x]
prime_list = prime_list.remove(2)
```

```python
limit = 100
res = (limit/2)**2
for i in prime_list:
    if i>limit:
        break
    #rint(i, numberToReduce(limit,i))
    res -= numberToReduce(limit,i)
for i1 in range(len(prime_list)):
    p1 = prime_list[i1]
    if p1>limit:
        break
    for i2 in range(i1+1,len(prime_list)):
        p2 = prime_list[i2]
        if p1*p2>limit:
            break
        #rint(p1*p2, numberToIncrease(limit,p1*p2))
        res += numberToReduce(limit,p1*p2)
res # scheint vom prinzip zu funktionieren
```
